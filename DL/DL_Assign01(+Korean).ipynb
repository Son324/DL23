**def tts(text):**

3단계. tts(text)를 정의하세요.

def tts(text): 이 줄은 텍스트 인수를 받는 tts 함수를 정의합니다. 이 함수의 목적은 주어진 텍스트를 오디오 파일로 변환하는 것입니다.

gtts_object = gTTS(text, lang = "en", slow - True): 이 줄은 주어진 텍스트를 사용하여 새로운 gTTS 물체를 생성하며, 언어를 영어로, 음성 속도를 느리게 명시합니다.

gtts_object.save("myaudio.mp3"): 이 줄은 gTTS 물체가 생성한 오디오 파일을 로컬 파일 시스템의 "myaudio.mp3" 파일에 저장합니다.

return Audio("myaudio.mp3"): 이 줄은 좀 전에 저장된 오디오 파일을 나타내는 Audio 물체를 돌려줍니다. 그 Audio 물체는 오디오 파일을 재생하거나 다양한 방식으로 그것을 조작하는데 사용될 수 있습니다.

```python
def tts(text):
  gtts_object = gTTS(text,lang = "en", slow = True) 
  gtts_object.save("myaudio.mp3")
  return Audio("myaudio.mp3")
```

# Using **input( )**

- In Python, input() is a built-in function that is used to get input from the user via the console. The input() function displays a prompt on the console, waits for the user to enter some text, and then returns the text entered by the user as a string.

- The syntax of the input() function is simple. It takes an optional string argument that is displayed as a prompt to the user, and then waits for the user to enter some text. 

input() 사용: 파이썬에서, input()은 콘솔을 통해 사용자로부터 입력을 받는데 사용되는 내장 함수입니다. 그 input() 함수는 콘솔에 프롬프트를 표시하고, 사용자가 몇몇의 텍스트를 입력하도록 기다리고, 그리고 그 후 사용자에 의해 입력된 텍스트를 문자열로 바꿉니다. input() 함수의 구문은 간단합니다. 사용자에게 프롬프트로서 나타내는 선택적인 문자열 인수를 취하고, 그리고 사용자가 몇몇의 텍스트를 입력하는 것을 기다립니다.
